# imdb
https://www.imdb.com/chart/top/

```python
import requests
from bs4 import BeautifulSoup
import html5lib

url = 'https://www.imdb.com/chart/top/'
html=requests.get(url).content
# requests.post(url).content
html
```

```python
soup=BeautifulSoup(html,"html5lib")
soup
```

```python
print(soup.title.text,"\n",70*'*',sep='')
# soup.title.text
```

```python
List =soup.find('tbody',{'class':'lister-list'}).find_all('tr')
# List
```

```python
for film in List:
    title = film.find('td',{'class':'titleColumn'}).a.text
    year = film.find('td',{'class':'titleColumn'}).span.text.strip('()')
    rating = film.find('td',{'class':'ratingColumn imdbRating'}).strong.text
    director = film.find('td',{'class':'titleColumn'}).a.get('title').split(' (dir.), ')[0]
    actors = film.find('td',{'class':'titleColumn'}).a.get('title').split(' (dir.), ')[1]
        
    show=f'''Title\t: {title}
Year\t: {year}
Rating\t: {rating}
Director: {director}
Actors\t: {actors}'''
    print(show)
    print(70*"*")
```

```python
title,year,rating,director,actors =[],[],[],[],[]
for film in List:
    title.append( film.find('td',{'class':'titleColumn'}).a.text )
    year.append( film.find('td',{'class':'titleColumn'}).span.text.strip('()') )
    rating.append( film.find('td',{'class':'ratingColumn imdbRating'}).strong.text )
    director.append( film.find('td',{'class':'titleColumn'}).a.get('title').split(' (dir.), ')[0] )
    actors.append(film.find('td',{'class':'titleColumn'}).a.get('title').split(' (dir.), ')[1] )

keys=['title','year','rating','director','actors']
values=[title,year,rating,director,actors]
import pandas as pd

pd.DataFrame(dict(zip(keys, values)))
```

# mediamarkt
https://www.mediamarkt.nl/nl/category/_laptops-482723.html

```python
import requests
from bs4 import BeautifulSoup
import html5lib

url=f"https://www.mediamarkt.nl/nl/category/_laptops-482723.html?searchParams=&sort=&view=&page=1"
html=requests.get(url).content
soup=BeautifulSoup(html,"html.parser")
print(soup.title.text,"\n",70*'*',sep='')

for i in range (1,5):
    url=f"https://www.mediamarkt.nl/nl/category/_laptops-482723.html?searchParams=&sort=&view=&page={i}"
    html=requests.get(url).content
    soup=BeautifulSoup(html,"html.parser")
#     print(soup.title.text,"\n",70*'*',sep='')

    liste=soup.find_all("div",{"class":"product-wrapper"})

    for i in liste:
        price=i.find("div",{"class":"price"}).text.strip("-").strip(",")
        comp_name=i.h2.text.strip()
        details_title =i.find("dl",{"class":"product-details"}).find_all("dt")
        details =i.find("dl",{"class":"product-details"}).find_all("dd")

        show=f'Name\t: {comp_name} \nPrice\t: {price} \nDetails\t: '    
        print(show, end='')
        for s1,s2 in zip(details_title,details):
            print(s1.text, s2.text,'\n\t  ',end='')
        print("\n",70*"*","\n")
```
